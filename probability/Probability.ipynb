```python
import numpy as np
import matplotlib.pyplot as plt

# Simulate all possible outcomes of rolling two dice
outcomes = np.arange(2, 13)  # Possible sums range from 2 to 12
# probabilities = [1/36, 2/36, 3/36, 4/36, 5/36, 6/36, 5/36, 4/36, 3/36, 2/36, 1/36]# this is not best method  because we use numpy
probabilities = np.array([1,2,3,4,5,6,5,4,3,2,1])/36

# Create a bar plot
plt.scatter(outcomes, probabilities, color='b', alpha=0.7)
plt.xlabel("Sum of Dice Rolls")
plt.ylabel("Probability")
plt.title("Probability Distribution of Sums of Two Dice Rolls")
plt.xticks(np.arange(2, 13)) 
plt.grid(axis='y', alpha=0.75)

plt.show()
```

```python
import matplotlib.pyplot as plt

# Calculate probabilities for each sum
probabilities = [0] * 11  # Initialize list to store probabilities
total_outcomes = 0

for die1 in range(1, 7):
    for die2 in range(1, 7):
        total_outcomes += 1
        probabilities[die1 + die2 - 2] += 1

# Normalize probabilities
probabilities = [prob / total_outcomes for prob in probabilities]

# Create a bar plot
outcomes = range(2, 13)
plt.bar(outcomes, probabilities, color='b', alpha=0.7 , label='PMF')
plt.xlabel("Sum of Dice Rolls")
plt.ylabel("Probability")
plt.title("Probability Distribution of Sums of Two Dice Rolls")
plt.xticks(outcomes)
plt.grid(axis='y', alpha=0.75)
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the possible sums and their corresponding probabilities
# possible_sums = list(range(2, 13))  # this is not best method , best method using numpy
possible_sums = np.arange(2,13)
# probabilities = [1/36, 2/36, 3/36, 4/36, 5/36, 6/36, 5/36, 4/36, 3/36, 2/36, 1/36] # this is not best method , best method using numpy
probabilities = np.array([1, 2, 3, 4, 5, 6, 5, 4, 3, 2, 1]) / 36
# Calculate cumulative probabilities
# cumulative_probabilities = [sum(probabilities[:i+1]) for i in range(len(possible_sums))]  # this is not best method 
#  if you remove numpy then all those line work that commented by line this is not best method
cumulative_probabilities = np.cumsum(probabilities)

# Create the plot
plt.figure(figsize=(10, 6))

# Cumulative Probability Plot
plt.step(possible_sums, cumulative_probabilities, where='mid', label='Cumulative Probability', color='blue')

# Probability Plot
plt.bar(possible_sums, probabilities, alpha=0.5, label='Probability', color='gray')

# Adding labels and title
plt.xlabel('Sum of Two Dice')
plt.ylabel('Probability')
plt.title('Cumulative Probability Distribution of Throwing Two Dice')
plt.legend()
plt.xticks(np.arange(2,13))
# Display the plot
plt.grid()
plt.show()
```

```python
#Geometric distribution with shoe mean and medain
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
from scipy.stats import geom
p = 0.4
x = np.arange(1,17)
y = stats.geom.pmf(x,p)
mean = geom.mean(p)
median = geom.median(p)
# Plotting
plt.figure(figsize=(10,7))
plt.xlabel("Probability of success after which point")
plt.ylabel("Probability of success")
plt.title("Geometric distribution")
plt.bar(x,y)
plt.axvline(mean, color='r', linestyle='dashed', label='mean')
plt.axvline(median, color='g', linestyle='dashed', label='mean')
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy import stats
from scipy.stats import geom
p_value =[0.4,0.2]
x = np.arange(1,21)
fig, axes = plt.subplots(ncols = 2 , nrows = len(p_value), figsize = (10, 6*len(p_value)))
for i, p in enumerate(p_value):
    pmf = stats.geom.pmf(x,p)
    cdf = geom.cdf(x,p)
    mean = geom.mean(p)
    median = geom.median(p)
#     for plot PMF
    axes[i,0].bar(x,pmf , label='PMF', color='yellow')
    axes[i,0].set_xlabel("X")
    axes[i,0].set_ylabel("Probability of outcomes")
    axes[i,0].set_title("Probability mass function Graph")
    axes[i,0].axvline(mean ,color='r',alpha = 0.7 , linestyle = 'dashed',label= 'mean')
    axes[i,0].axvline(median ,color='gray',alpha = 0.7 , linestyle = 'dashed' , label = 'median')
    axes[i, 0].legend()  # from this line label will occur
    
#      for plot CDF
    axes[i,1].step(x,cdf , label='CDF', color='red')
    axes[i,1].axvline(median, color='g', linestyle='dashed', label='Median')
    axes[i,1].axhline(0.5 , color='b' , linestyle = 'dashed',label= 'f(x) = 0.5' )
    axes[i,1].legend()
plt.tight_layout()
plt.show()    
```
