```python
# Q-2
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

n_sample = 50
n_population = 5000

# Generate right-skewed data with α = 1, β = 5
alpha_right_skewed = 1
beta_right_skewed = 5
data_right_skewed = np.random.beta(alpha_right_skewed, beta_right_skewed, n_sample)

# Generate symmetric data with α = 5, β = 5
alpha_symmetric = 5
beta_symmetric = 5
data_symmetric = np.random.beta(alpha_symmetric, beta_symmetric, n_sample)

# Generate left-skewed data with α = 5, β = 1
alpha_left_skewed = 5
beta_left_skewed = 1
data_left_skewed = np.random.beta(alpha_left_skewed, beta_left_skewed, n_sample)

# Calculate summary statistics
def summarize_data(data, label):
    mean = np.mean(data)
    median = np.median(data)
    mode = stats.mode(data).mode.item()
    print(label)
    print("Mean:", mean)
    print("Median:", median)
    print("Mode:", mode)

# Summarize each distribution
summarize_data(data_right_skewed, "Right-Skewed Data (α = 1, β = 5)")
summarize_data(data_symmetric, "Symmetric Data (α = 5, β = 5)")
summarize_data(data_left_skewed, "Left-Skewed Data (α = 5, β = 1)")

# Plot histograms
plt.figure(figsize=(12, 4))
plt.subplot(1, 3, 1)
plt.hist(data_right_skewed, bins=20)
plt.title("Right-Skewed Data")

plt.subplot(1, 3, 2)
plt.hist(data_symmetric, bins=20)
plt.title("Symmetric Data")

plt.subplot(1, 3, 3)
plt.hist(data_left_skewed, bins=20)
plt.title("Left-Skewed Data")
plt.tight_layout()
plt.show()

# Plot box-and-whisker plots
plt.figure(figsize=(12, 4))
plt.subplot(1, 3, 1)
plt.boxplot(data_right_skewed)
plt.title("Box-and-Whisker Plot - Right-Skewed Data")

plt.subplot(1, 3, 2)
plt.boxplot(data_symmetric)
plt.title("Box-and-Whisker Plot - Symmetric Data")

plt.subplot(1, 3, 3)
plt.boxplot(data_left_skewed)
plt.title("Box-and-Whisker Plot - Left-Skewed Data")
plt.tight_layout()
plt.show()
```

```python
# Q-3
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

n_population = 5000
n_sample = 50

# Generate right-skewed data with α = 1, β = 5
alpha_values = [1, 2, 5]
beta_values = [0.5, 1, 2]
for alpha,beta in zip(alpha_values,beta_values):
    data_skewed = np.random.gamma(alpha, beta, n_population)
    
    plt.figure(figsize=(10, 7))
    plt.subplot(1, 2, 1)
    sns.histplot(data=data_skewed, kde=True)
    plt.title(f"Histogram and KDE Plot - α = {alpha}, β = {beta}")
    
    plt.subplot(1, 2, 2)
    sns.boxplot(data=data_skewed)
    plt.title(f"Box Plot - α = {alpha}, β = {beta}")
    
    plt.tight_layout()
    plt.show()
```

```python
# Q-3
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

n_population = 5000
n_sample = 50

# Generate random alpha and beta values within specified ranges
alpha_values = np.random.uniform(0.5, 5, n_sample)
beta_values = np.random.uniform(0.5, 5, n_sample)

for alpha, beta in zip(alpha_values, beta_values):
    data_skewed = np.random.gamma(alpha, beta, n_population)
    
    plt.figure(figsize=(10, 7))
    plt.subplot(1, 2, 1)
    sns.histplot(data=data_skewed, kde=True)
    plt.title(f"Histogram and KDE Plot - α = {alpha:.2f}, β = {beta:.2f}")
    
    plt.subplot(1, 2, 2)
    sns.boxplot(data=data_skewed)
    plt.title(f"Box Plot - α = {alpha:.2f}, β = {beta:.2f}")
    
    plt.tight_layout()
    plt.show()
```

```python
# Q-10
from scipy import stats
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

nsample = 200
mu, sigma = 10, 2
n = 10
Z = np.zeros(nsample, float)

for i in range(nsample):
    X = stats.norm.rvs(mu, sigma, n)
    X_df = pd.DataFrame({"X": X})
    mean = X_df.mean(axis=0)
    Z[i] = (mean - mu) / (sigma / np.sqrt(n))

Z_df = pd.DataFrame({"Z": Z})
print(Z_df.describe())

fig = plt.figure(figsize=(10, 10))
ax1 = fig.add_subplot(2, 1, 1)
ax2 = fig.add_subplot(2, 1, 2)

sns.histplot(Z_df["Z"], stat="density", alpha=0.3, ax=ax1)
y = np.linspace(-3, 3, 100)
sns.lineplot(x=y, y=stats.norm.pdf(y), color='r', ax=ax1)

res = stats.probplot(Z, plot=ax2)
plt.show()
```
